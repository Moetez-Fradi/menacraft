# Threshold and Calibration Evaluation
Evaluate score thresholds and confidence calibration.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score

df = pd.read_csv('../data/calibration_scores.csv')
y_true = df['label'].values
scores = df['ai_generated_score'].values

best_t, best_f1 = 0.5, 0.0
for t in np.linspace(0.1, 0.9, 17):
    y_pred = (scores >= t).astype(int)
    f1 = f1_score(y_true, y_pred)
    if f1 > best_f1:
        best_t, best_f1 = t, f1

print({'best_threshold': best_t, 'best_f1': best_f1})
```
