# Build Reference Embeddings
Seed Qdrant collections with local reference samples for explainability and retrieval.

```python
from pathlib import Path
from app.embedders.embeddings import EmbedderService
from app.qdrant.store import QdrantEvidenceStore
from app.shared.constants import TEXT_EMBEDDINGS_COLLECTION

embedder = EmbedderService()
qdrant = QdrantEvidenceStore()
samples = [
    ("real_text", "This is a real journalistic paragraph about a local event."),
    ("ai_text", "In conclusion, this comprehensive analysis underscores the transformative impact..."),
]

for idx, (label, text) in enumerate(samples, start=1):
    vec = embedder.embed_text(text)
    qdrant.upsert_vector(TEXT_EMBEDDINGS_COLLECTION, idx, vec, {"label": label, "modality": "text", "artifact_id": f"ref-text-{idx}"})

print("Seed complete")
```
